## Сортировка пузырьком (bubble sort)



```python
arr = [25, 21, 22, 24, 23, 27, 26]
N = len(arr)

print(arr)
 
for i in range(N-1, 0, -1):
    for j in range(i):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]

print("\nSorted array")
print(arr)
```

### Сортировка вставками (insertion sort)

```python
def insertion_sort(arr):
    N = len(arr)
    
    for i in range(1, N):
        j = i - 1
        temp = arr[i]
        
        while arr[j] > temp and j >= 0:
            arr[j+1] = arr[j]
            j -= 1
        
        arr[j+1] = temp
    
    return arr
```

```python
A = [25, 26, 22, 24, 27, 23, 21]

print(f"Source array:\n{A}")

insertion_sort(A)

print(f"\nSorted array array:\n{A}")
```
