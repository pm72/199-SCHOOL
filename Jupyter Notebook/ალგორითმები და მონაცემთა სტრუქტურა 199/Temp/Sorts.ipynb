```python
def bubble_sort(A):
  '''Bubble sort'''
  
  N = len(A)
  
  for j in range(N - 1, 0, -1):
    for i in range(j):
      if A[i] > A[i+1]:
        A[i], A[i+1] = A[i+1], A[i]
  
  
def insertion_sort(A):
  '''Insertion sort'''
  
  N = len(A)
  
  for i in range(1, N):
    j = i
    while j > 0 and A[j-1] > A[j]:
      A[j], A[j-1] = A[j-1], A[j]
      j -= 1
  
  
def selection_sort(A):
  '''Selection sort'''
  
  N = len(A)
  
  for i in range(N - 1):
    for j in range(i + 1, N):
      if A[i] > A[j]:
        A[i], A[j] = A[j], A[i]

        
def shell_sort(A):
  '''Shell sort'''
  
  N = len(A)
  distance = N // 2
  
  while distance > 0:
    for i in range(distance, N):
      j = i
      while j >= distance and A[j] < A[j-distance]:
        A[j], A[j-distance] = A[j-distance] = A[j]
        j -= distance
    
    distance //= 2
```

```python
from random import randint

def test_sorts(sort_algorithm):
  print("Testing", sort_algorithm.__doc__)
  
  # test case #1
  print("Test case #1: ", end='')
  
  arr = [4, 2, 5, 1, 3]
  arr_sorted = [1, 2, 3, 4, 5]
  
  sort_algorithm(arr)
  
  print("OK!" if arr == arr_sorted else "FAIL!!!")
  
  
  # =========================
  # test case #2
  print("Test case #2: ", end='')
  
  arr = [randint(1, 100) for i in range(20)]
  arr_sorted = sorted(arr)
  
  sort_algorithm(arr)
  
  print("OK!" if arr == arr_sorted else "FAIL!!!")
  
  
  # =========================
  # test case #3
  print("Test case #3: ", end='')
  
  arr = [4, 2, 4, 1, 3, 2]
  arr_sorted = [1, 2, 2, 3, 4, 4]
  
  sort_algorithm(arr)
  
  print("OK!" if arr == arr_sorted else "FAIL!!!")
```

```python
if __name__ == '__main__':
  test_sorts(bubble_sort)
  print()
  
  test_sorts(insertion_sort)
  print()
  
  test_sorts(selection_sort)
  print()
  
  test_sorts(shell_sort)
  print()
```
