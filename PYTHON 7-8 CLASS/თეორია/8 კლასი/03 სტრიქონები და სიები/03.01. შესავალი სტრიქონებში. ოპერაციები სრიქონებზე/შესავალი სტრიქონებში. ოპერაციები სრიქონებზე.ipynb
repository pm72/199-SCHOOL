### ოპერატორი `in`

```python
print("ata" in "Paata Mamporia")
print("ata Ma" in "Paata Mamporia")
print("ata ma" in "Paata Mamporia")
```

### სტრიქონების შედარება

```python
a = "Paata Mamporia"
print(a == "Paata Mamporia")
print(a == "paata mamporia")
print(a == "Hello")

print()

print(a != "Hello")   
```

### სტრიქონების ლექსიკოგრაფიკული შედარება

```python
# სიმბოლ-სიმბოლ შედარება. სიმბოლოების რეგისტრი მხედველობაში მიიღება
print("cat" > "cet")
print("cat" > "cat")
print("cat" == "cat")
print("cat" >= "cat")
print("cat" <= "cat")

print()

print("cat" == "Cat")
print("cat" > "Cat")
print("cat" < "Cat")
```
