#### მეთოდი `count()`

`String.count(sub[start, [end]])`

`start` – ინდექსი, საიდანაც იწყება ძებნა

`end` – ინდექსი, სადაც მთავრდება ძებნა

`[, ]` – არასავალდებულო

````python
msg = "abrakadabra"

print(msg.count('a'))
print(msg.count('ra'))
````

````python
msg = "abrakadabra"

print(msg.count('ra', 4))
print(msg.count('ra', 4, 10))
print(msg.count('ra', 4, 11))
````

#### მეთოდები: `find()`, `rfind()`, `index()`

`String.find(sub[start, [end]])`

`start` – ინდექსი, საიდანაც იწყება ძებნა

`end` – ინდექსი, სადაც მთავრდება ძებნა

`[, ]` – არასავალდებულო

````python
msg = "abrakadabra"

print(msg.find("br"))
print(msg.find("br", 2))
print(msg.find("brr"))
````

მეთოდი `find()` ახორციელებს ძებნას მარცხნიდან მარჯვნივ.

მეთოდი `rfind()` ანალოგიურია წინა მეთოდისა იმ განსხვავებით, რომ ძებნა წარმოებს მარჯვნიდან მარცხნივ.

````python
msg = "abrakadabra"

print(msg.rfind("br"))
````

მეთოდი `index()` ანალოგიურია `find()` მეთოდისა, ოღონდ თუ საძიებო სტრიქონი ვერ იპოვა, ეკრანზე გამოიტანს შეცდომას.

````python
msg = "abrakadabra"

print(msg.index('br'))
print(msg.index('brr'))
````

#### მეთოდი `replace()`

`String.replace(old, new, count=-1)`

`count` – ცვლილების მაქსიმალური რაოდენობა (-1 – შეუზღუდავი)

````python
msg = "abrakadabra"

print(msg.replace("a", "o"))
print(msg.replace("ab", "AB"))
print(msg.replace("ab", ""))

print()

print(msg.replace('a', 'o', 2))
````

#### მეთოდი `isalpha()`

მეთოდი აბრუნებს `True`-ს, თუ სტრიქონი მხოლოდ ასოებისგან შედგება, წინააღმდეგ შემთხვევაში – `Flase`-ს.

````python
msg = "abrakadabra"

print(msg.isalpha())
print("Hello world".isalpha())
print("Hello,".isalpha())
````

#### მეთოდი `isdigit()`

მეთოდი აბრუნებს `True`-ს, თუ სტრიქონი მხოლოდ сшакуишыпфт შედგება, წინააღმდეგ შემთხვევაში – `Flase`-ს.

````python
print("5.6".isdigit())
print("56".isdigit())
print("-56".isdigit())
````

#### მეთოდი `rjust()`

`String.rjust(width [, filchar=''])`

````python
d = "abc"
print(d.rjust(5))

d = "12"
print(d.rjust(4))
print(d.rjust(4, '0'))
print(d.rjust(1, '0'))
````

#### მეთოდი `ljust()`

`String.ljust(width [, filchar=''])`

````python
d = "12"
print(d.ljust(10, '*'))
````

#### მეთოდი `center()`

`String.ljust(width [, filchar=''])`

````python
d = "String methods"
print(d.center(50))
print(d.center(50, '*'))
````

#### მეთოდი `split()`

`String.split(sep=None, maxsplit=-1)`

`sep` – სტრიქონის დამყოფი სიმბოლო(ები) (დუმლის პრინციპით: `' ', '\n', '\t'`)

`maxsplit` – სტრიქონის დაყოფის რაოდენობა (-1 – შუზღუდავი)

````python
msg = "I love Python very much"
print(msg.split(" "))
print(msg.split())

msg = "I love Python\nvery much"
print(msg.split())

msg = "I love Python\nvery\t\tmuch"
print(msg.split())
````

````python
# უნდა დავბეჭდოთ: 1, 2, 3, 4, 5, 6

digs = "1, 2,3, 4, 5,6"
print(digs.split(','))
print(digs.split(', '))
print(digs.replace(' ', ''))
print(digs.replace(' ', '').split(','))

print()

print(*digs.replace(' ', '').split(','))
print(*digs.replace(' ', '').split(','), sep=', ')
````

#### მეთოდი `join()`

`String.join(list)`

````python
digs = "1, 2,3, 4, 5,6"
d = digs.replace(' ', '').split(',')
s = ', '.join(d)

print(s)
print(', '.join(digs.replace(' ', '').split(',')))
````

````python
info = "Paata Mamporia 51 Tbilisi Georgia"
info = ', '.join(info.split()) 
print(info)
````

#### მეთოდები strip(), lstrip(), rstrip()

მეთოდი მთლიან სტრიქონს მარჯვნიდან და მარცხნიდან აცილებს ზედმეტ ინტერვალს, ახალ ხაზზე გადასვლის სიმბოლოს (`\n`), ჰორიზონტალურ ტაბულაციას (`\t`).

`lstrip()` და `rstrip()` იგივეა რაც `strip()`, ოღონდ არასასურველ ინტერვალებს აცილებს შესაბამისად მხოლოდ მარცხენა ან მარჯვენა მხარეს.

`String.strip()`

`String.lstrip()`

`String.rstrip()`

````python
s = "    \t\t   Hello, world!..      \n"
print(s)
````

````python
print(s.strip())
````

````python
print(s.lstrip())
````

````python
print(s.rstrip())
````
