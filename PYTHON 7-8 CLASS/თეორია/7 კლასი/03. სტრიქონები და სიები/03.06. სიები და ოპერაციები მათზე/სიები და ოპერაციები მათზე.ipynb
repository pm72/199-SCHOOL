სტრიქონებისგან განსხვავებით სია ცვლად ტიპს მიეკუთვნება: შესაძლებელია სიის ელემენტების შეცვლა (შესაბამისი ინდექსით სიის ელემენტზე სრულად წვდომა).

#### სიაში ჩაშენებული ფუნქციები

```python
len() – სიის სიგრძე, ელემენტების რაოდენობა
max() – მაქსიმალური ელემენტის პოვნა
min() – მინიმალური ელემენტის პოვნა
sum() – ჯამის გამოთვლა
sorted() – კოლექციის სორტირება
```

````python
marks = [8, 'ფრიადი', 10, 8, 8, 9]

print(len(marks))
print(len([]))
````

````python
t = [33.5, 35.6, 37.3, 36.0, 40.0, 39.5]

print(t)
print(min(t), max(t))
print(sum(t))

print(f"ტემპერატურის საშუალო მაჩვენებელი: {sum(t) / len(t)}")

print(f"სორტირებული ტემპერატურა ზრდადობის მიხედვით: {sorted(t)}")

t_sort = sorted(t)
print(t_sort)
print(t, "\n")

print("სორტირება კლაბადობის მიხედვით")
t_sort = sorted(t, reverse=True)
print(t_sort)
````

`min(), max(), sorted()` ფუნქციები ნებისმიერ ობიექტთან მუშაობენ, სადაც შესაძლებელია შედარება და სორტირება. მაგალითისთვის ავიღოთ სტრიქონები.

````python
s = list('python')

print(max(s))
print(min(s))
print(sorted(s))
print(sum(s))
````

`+` – ორი სიის გაერთიანება ერთ სიად  
`*` – სიის დუბლირება  
`in` – სიაში ელემენტის არსებობის შემოწმება  
`del` – სიიდან ელემენტის წაშლა  

````python
a = [1, 2, 3] + [4, 5]

print(a)

# [1, 2, 3] + 4  არ შეიძლება: აღიძვრება შეცდომა.
# + ოპერატორი ამ შემთხვევაში მხოლოდ სიების გაერთიანებისთვის გამოიყენება
# [1, 2, 3] + [4] + ["Good"] – შესაძლებელია

[1, 2, 3] + [4] + ["Good"]
````

````python
a = ["I", "love", "python"]
print(a * 3) # აუცილებლად მთელი რიცხვი
print(a + [3])
print(a + 3)
````

````python
["I"] + ["love"] * 3 + ["python"]
````

````python
# in – სიაში ელემენტის არსებობის შემოწმება. შედეგი: True / False

lst = ["თბილისი", 1320, 5.8, True, "გლდანი", False]

print(1320 in lst)
print(132 in lst)
print([1, 3] in lst)
print("თბილისი" in lst)
````

````python
# del – ელემენტის წაშლა სიიდან

lst = ["თბილისი", 1320, 5.8, True, "გლდანი", False, [1, 3]]

del lst[2]

print(lst)
````
