# რიცხვი და არითმეტიკული ოპერაციები

* `int` – მთელი ტიპი
* `float` – მცოცავწერტილიანი (ათწილადი) ტიპი
* `complex` – კომპლექსური ტიპი


__მთელი რიცხვები:__ `0, 1, 2, 15872457812011, -1, -2, -784125`

__მცოცავწერტილიანი (ათწილადი) რიცხვები:__ `6.8, -1.24, 345.214, -6587.26`

არითმეტიკული ოპერაციები: `+`, `-`, `*`, `/, //`, `%`, `**`

#### Shell

```python
print(2 + 3)

a = 2 + 3
print(a)

b = 2 + 3 + 4
print(b)

c = 2 + 3.5
print(c)

c = 2 + 3.0
print(c)

d1 = 8 / 2
print(d1)

d2 = 3 / 6
print(d2)

d3 = 7 // 2  # მთელზე გაყოფა (დამრგვალება ნაკლებ მთელ რიცხვამდე)
print(d3)

print(-7 // 2)
print(-7 // -2)

print(6 * 5)
print(2 * 4.5)
```

```python
# ნაშთი (%)
print(10 % 3)
print(10 % 4)
```

```python
print(-9 % 5)
```

```python
print(9 % -5)
```

```python
print(-9 % -5)
```

```python
# ხარისხში აყვანა (**)

print(2 ** 3)
print(6 ** 2)
print(36 ** 0.5)
print(2 ** 3 ** 2)
```

```python
# არითმეტიკული ოპერაციების პრიორიტეტი

print(27 ** 1/3)
print(27 ** (1/3))

print()

print(2 + 3 * 5)
print((2 + 3) * 5)

print()

print(32 / 4 * 2)
```

```python
i = 5
j = 3

i = i + 1
print(i)

j = j - 2
print(j)

print()

i += 1   # i = i + 1
print(i)

j -= 2  # j = j - 2
print(j)

print()

i *= 3
print(i)

j /= 4
print(j)
```
