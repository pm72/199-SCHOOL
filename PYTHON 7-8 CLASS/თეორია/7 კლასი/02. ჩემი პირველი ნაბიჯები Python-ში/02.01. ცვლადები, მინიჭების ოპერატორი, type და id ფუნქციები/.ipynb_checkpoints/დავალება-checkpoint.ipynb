__დავალება 3.__

რა ხდება ერთი ცვლადის მეორე ცვლადზე მინიჭებისას (`a = b`)?

* ასეთი ოპერაცია დაუშვებელია. შესაძლებელია მხოლოდ ობიექტების მინიჭება და არა ცვლადების.
* მონაცემთა კოპირება და ცვლადი `a` შეიცავს ინფორმაციის ასლს `b` ცვლადიდან
* ბმულის კოპირება და ორივე ცვლადი მიუთითებს ერთსა და იმევე ობიექტს
* იქმნება ორი ცვლადი, თუ ისინი აქამდე არ არსებობდა, მიენიჭებათ ნულოვანი მნიშვნელობა

__დავალება 10.__

რა მოხდება, თუ ფუნქცია `type`-ს მივანიჭებთ რიცხვს `type = 7`?

* დაიბეჭდება შეცდომა: `type` – ჩანერგილი ფუნქციაა `Python`-ში
* ცვლადი `type` მიუთითებს რიცხვს `7`-ს
* ფუნქცია `type` დააბრუნებს `7`-ს
* ინტერპრეტატორი ოპერაციას მხედველობაში არ მიიღებს და პროგრამა გააგრძელებს მუშაობას
* რიცხვი `7` გახდება `type` ფუნქციის სინონიმი
