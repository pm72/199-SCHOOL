# ჭრა. შედარების ოპერატორები

```python
lst = ["რუსთავი", "თელავი", "სოხუმი", "გორი"]

print(lst[1:3])

a = lst[1:3]

print(id(lst))
print(id(a))

a[0] = "ლაგოდეხი"
print(a)
print(lst)
```

```python
lst = ["რუსთავი", "თელავი", "სოხუმი", "გორი"]

print(lst[2:])  # 2-დან ბოლომდე. მხედველობაში მიიღება ინდექსი
print(lst[:3], "\n")  # თავიდან 2-ის ჩათვლით. მხედველობაში მიიღება ინდექსი

# სიის კოპირება
cities = lst[:] 
print(id(lst))
print(id(cities), "\n")

c = list(lst)
print(id(lst))
print(id(c), "\n")

d = lst.copy()
print(id(lst))
print(id(d))
```

```python
lst = ["რუსთავი", "თელავი", "სოხუმი", "გორი"]

a = lst  # არ ნიშნავს სიის კოპირებას.
# a და lst ცვლადები მიუთითებენ ერთსა და იმევაე ობიექტს (ამ შემთხვევაში სიას).

print(id(lst))
print(id(a), "\n")

a[0] = 158
print(lst)
print(a, "\n")
```

```python
# უარყოფითი ინდექსი ===> სიის ბოლო ელემნტი
lst = ["რუსთავი", "თელავი", "სოხუმი", "გორი"]
print(lst[-1])
print(lst[-2])
print(lst[-5])
```

```python
marks = [4, 6, 8, 6, 10, 4]

print(marks[2:-1])  # მეორედან ბოლოსწინის ჩათვლით
print(marks[-3:-1])
```

```python
marks = [4, 6, 8, 6, 10, 4]

print(marks[1:5:2])
print(marks[:5:2])
print(marks[1::2])
print(marks[::2], "\n")

print(marks[::-1])  # შებრუნებული სია (reverse)
print(marks[::-2])
```

#### ელემენტების ჯგუფის შეცვლა

```python
marks = [4, 6, 8, 6, 10, 4]

marks[2:4] = ["კარგი", "დამაკმაყოფილებელი"]
print(marks, "\n")

marks[::2] = [0, 0, 0]
print(marks, "\n")

marks += [6]
print(marks)
marks[::2] = [0, 0, 0]
```

```python
marks[:5:2] = [0, 0, 0]
print(marks, "\n")

marks[2:4] = 10, 20
print(marks)
```

  ### სიების შედარება <, >, <=, >=, ==, !=

```python
print([1, 2, 3] == [1, 2, 3])
print([1, 2, 3] != [1, 2, 3])
print([1, 2, 3] < [1, 2, 3])
print([1, 2, 3] > [1, 2, 3])
print([1, 2, 3] <= [1, 2, 3])
print([1, 2, 3] >= [1, 2, 3], "\n")

print([10, 2, 3] > [1, 2, 3])
print([10, 2, 3] > [1, 20, 3])
print([1, 2, 3, 4] > [1, 2, 3])
print([1, 2, 3, 4] > [10, 2, 3], "\n")

print([1, 2, "abc"] > [1, 2, "abc"])
print([1, 2, 3] > [1, 2, "abc"])
# პითონს არ შეუძლია შეადორას რიცხვი სტრიქონს. ის ადარებს რიცხვს რიცხვთან ან
# სტრიქონს სტრიქონთან.
```
