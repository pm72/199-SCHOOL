# მეთოდები

```python
a = [1, -56, 3, 23, 43, -45, 0]

# ელემენტის სიის ბოლოში დამატება .append()
a.append(100)
print(a, "\n")

# არ გააკეთოთ ასე
a = a.append(40)
print(a)
```

```python
a = [1, -56, 3, 23, 43, -45, 0]

a.append(True)
print(a, "\n")

a.append([1, 2, 3])
print(a, "\n")

a.append(51, 87)
```

```python
# ელემენტის სიაში ჩასმა მითითებულ ადგილას .insert(i, item)
# i – ინდექსი, პოზიცია; item – ჩასასმელი ელემენტი
a = [1, -56, 3, 23, 43, -45, 0, True, [1, 2, 3]]

a.insert(3, -1000)
print(a, "\n")
```

```python
# ელემენტის სიიდან წაშლა .remove(item)
# item – სიიდან წასაშლელი ელემენტი
a = [1, -56, 3, -1000, 23, 43, -45, 0, True, [1, 2, 3]]

a.remove(True)   # True = 1   False = 0
print(a, "\n")

a.remove(True)
print(a, "\n")

a.remove("Hello")
print(a)
```

```python
# ელემენტის სიიდან წაშლა .pop(), .pop(i)
# i – ინდექსი, პოზიცია
# .pop() მეთოდი ბოლო ელემნტს შლის სიიდან
# წაშლილი ელემენტის დამახსოვრება შესაძლებელია

a = [1, -56, 3, -1000, 23, 43, -45, 0, True, [1, 2, 3]]

end = a.pop()
print(end)
print(a, "\n")

a.pop()
print(a, "\n")

a.pop(3)
print(a, "\n")
```

```python
# სიის გასუფთავება .clear()

a = [1, -56, 3, -1000, 23, 43, -45, 0, True, [1, 2, 3]]

a.clear()
print(a, "\n")
```

```python
# სიის კოპირება. მეთოდი .copy()

a = [1, -56, 3, 23, 43, -45, 0]

c = a.copy()   # c = a.copy()  ===>  c = a[:]  ===>  c = list(a)

print(id(a))
print(id(c))
```

```python
# სიაში არსებული ელემენტების რაოდენობა. მეთოდი .count(item)
# item – სიის ელემენტი 

a = [1, -56, 3, 23, 43, -45, 0]
c = a.copy()

print(c.count(1), "\n")

c.append(1)
print(c)
print(c.count(1), "\n")

print(c.count(True), "\n")

print(c.count("Hello"), "\n")
```

```python
# სიაში არსებული ელემენტის ინდექსი, პოზიცია. მეთოდი .index(item[, start])
# item – სიის ელემენტი; start – ძებნის დიაპაზონი, დუმილის პრინციპით start = 0

a = [1, -56, 3, 23, 43, -45, 0, 1]

print(a.index(1))
print(a.index(-56))
print(a.index(1, 2))
print(a.index("Hello"))
```

```python
"Hello" in  a

if "Hello" in a:
  print(a.index("Hello"))
else:
  print("'Hello' not in list")
```

```python
# სიის შებრუნება, რევერსი, სიის ინვერსია. მეთოდი .reverse()

a = [1, -56, 3, 23, 43, -45, 0]

a.reverse()
print(a, "\n")
```

```python
# სიის სორტირება. მეთოდი .sort([reverse=True])
# დუმილის პრინციპით reverse = False

a = [1, -56, 3, 23, 43, -45, 0, 1]

# სორტირება ზრდადობის (არაკლებადობის) მიხედვით
a.sort()
print(a)
print(sorted(a), "\n")
# .sort() მეთოდი „ადგილზე“ ცვლის სიას, არ აბრუნებს სიას,
# დუმილის პრინციპით აბრუნებს None-ს,
#ხოლო sorted() ფუნქცია აბრუნებს სორტირებულ, ახალ სიას.

# სორტირება კლებადობის (არაზრდადობის) მიხედვით
a.sort(reverse=True)
print(a)
print(sorted(a, reverse=True), "\n")
```

```python
cities = ["Tbilisi", "Moscow", "Auckland", "Sidnay", "Beijing", "Tokyo"]

cities.sort()
print(cities)
```
