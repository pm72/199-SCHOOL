# ჩადგმული სიები

```python
# შევქმნათ ერთგანზომილებიანი სია
line = [1, 7, 6, 11, 3]
# ეს ის მონაცემებია,რომლებიც აღწერს ილუსტრაციის პირველ სტრიქონს
  
# გამოსახულება ციფრების სახით შეიძლება გამოისახოს ჩადგმული სიების სახით,
# ჩვენ შემთხვევაში 5 სტრიქონი და 5 სვეტი (5x5)
img = [[1, 7, 6, 11, 3],
       [1, 7, 6, 11, 3],
       [1, 7, 6, 11, 3],
       [1, 7, 6, 11, 3],
       [1, 7, 6, 11, 3]
      ]
print(img)
```

```python
# შესაძლებელია img ცლადის უფრო მარტივად შექმნა
img = [line[:], line[:], line[:], line[:], line[:]]

print(img, "\n")

print(img[0], "\n")
```

```python
print(img[0])
print(img[0][0])
print(img[0][1])
print(img[0][2])
print(img[0][3])
print(img[0][4])
print(img[0][-1])
```

```python
img = [[1, 7, 6, 11, 3],
       [1, 7, 6, 11, 3],
       [1, 7, 6, 11, 3],
       [1, 7, 6, 11, 3],
       [1, 7, 6, 11, 3]
      ]

img[1] = [0, 0, 0, 0, 0]
print(img, "\n")

# img[1] = [0, 0, 0, 0, 0] უფრო მარტივად შეიძლება ჩავწეროთ
# img[1] = [1] * 0

# შესაძლებელია ჭრების მიხედვით შევცვალოთ მნიშვნელობები
img[1][:] = [1, 8, 96, 220, 178]
print(img, "\n")

img[1][:] = [1, 8, 96, 220, 178, 200, 254, 113]
print(img, "\n")

img[1][:] = [1] * 5
print(img, "\n")
```
