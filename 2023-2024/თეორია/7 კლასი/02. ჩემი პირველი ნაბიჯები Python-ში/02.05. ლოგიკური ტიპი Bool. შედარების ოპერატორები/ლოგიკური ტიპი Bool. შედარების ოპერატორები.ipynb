შედარების ოპერატორები: `<, >, <=, >=, ==, !=`

```python
print(4 > 2)
print(4 > 7)
```

```python
a = 5
b = 7.8

print(a <= b)
```

```python
a = 5
b = 7.8

res = a > b
print(res)
print(type(res))
```

```python
res = True
print(res)
```

```python
print(5 == 7 - 2)
print(8 >= 8)
print(8 > 8)
```

```python
# == შემოწმება ტოლობაზე

x = 6
print(x)

print(x == 6)
```

```python
# x ცვლდის შემოწმება ლუწობაზე (2, 4, 6, 8, ...)
x = 6
print(x % 2 == 0)

x = 7
print(x % 2 == 0)

print()

# x ცვლადის შემოწმებაკენტობაზე (1, 3, 5, 7, 9, ...)
print(x % 2 != 0)
print(x % 2 == 1)
```

```python
# არის თუ არა a ცვლადი 3-ის ჯერადი
a = 15
print(a % 3 == 0)

a = 16
print(a % 3 == 0)
```

```python
# შევამოწმოთ ცვლადი y მდებარეობს თუ არა [-2; 5] დიაპაზონში
y = 1.85
print(y >= -2 and y <= 5)
print(-2 <= y <= 5)
```

```python
# შევამოწმოთ ცვლადი y არ მდებარეობს [-2; 5] დიაპაზონში
y = 1.85
print(y < -2 or y > 2)

print()

print(y < -2)
print(y > 5)
```

```python
y = -10
print(y < -2 or y > 5)
```

```python
# ექვივალენტური გამოსახულებები

# x ცვლადის 2-ის და 3-ის ჯერადობაზე დადგენა 
x = 7

print(x % 2 == 0 and x % 3 == 0)         # False

print(x % 2 != 0 and x % 3 != 0)         # True

print(not (x % 2 == 0 and x % 3 == 0))   # True
```

### ინფორმაციის გარდაქმნა bool ტიპად

bool()   => True ან False

ნული, ცარიელი სტრიქონი ====> False (მცდარი)

ყველა, სხვა დანარჩენი (1, 2, 58, -2,5, ,0,05, ...) =====> True (ჭეშმარიტი)

```python
print(bool(1))
print(bool(-10))
print(bool(5.87))
print(bool("Hello"))
print(bool(" "))
```

```python
print(bool(0))
print(bool(0.0))
print(bool(""))
print(bool(''))
```
