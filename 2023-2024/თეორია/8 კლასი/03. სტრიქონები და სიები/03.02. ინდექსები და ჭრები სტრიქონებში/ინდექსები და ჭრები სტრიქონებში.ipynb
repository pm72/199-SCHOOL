როდესაც ვიყენებთ ინდექსს (`s[1], s[5], s[-2], ...`), ვღებულობთ ახალ სტრიქონს (ერთსიმბოლოიანი).

### სტრიქონის ჭრები (slice)

შესაძლებელია მრავალსიმბოლოიანი ახალი სტრიქონის მიღება ჩანაწერით.

`s[start:stop:step]`

ამას სტრიქონის ჭრა (slice) ეწოდება.

`s[start:stop:step]` ბოლო ორი პარამტერის მითითება აუცილებელი არაა.

````python
s = "Hello python"

print(s[1:3])
````

ამ შემთხვევაში იბეჭდება ახალი სტრიქონი ინდექსით 1-დან 3-მდე. 3 არ შედის ჭრის დიაპაზონში. დაიბეჭდება სიმბოლოები მდებარე 1 და 2 ინდექსზე.

````python
s = "Hello python"

print(s[4:])
print(s[:5])
print(s[:])

print()

a = s[:]

print(id(s))
print(id(a))
````

````python
s = "Hello python"

print(s[2:-2])
print(s[-2:2])
print(s[4:2])
````

````python
s = "Hello python"

print(s[2:10:2])
````

````python
s = "Hello python"

print(s[2::3])
````

````python
s = "Hello python"

print(s[:5:3])
print(s[::2])
print(s[::-1])
print(s[::-2])
````

სტრიქონი წარმოადგენს არაცვალებად მონაცემთა ტიპს. შექმნილი სტრიქონის შეცვლა არ შეიძლება,

````python
s = "hello python"

s[0] = 'H'
````

````python
s = "hello python"
print(id(s), s)

s = "H" + s[1:]

print(id(s), s)
````

* სტრიქონიდან ქვესტრიქონის (გარკვეული სიმბოლოების) მიღება შესაძლებელია ჭრების (slice) საშუალებით
* სტრიქონი არაცვალვადი ტიპის მონაცემთა სტრუქტურაა
